```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('StudentsPerformance.csv')
```

```python
data.describe()
```

```python
#univariant countplot:
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='gender', palette='pastel')
plt.title("Count of Male And Female")
plt.show()
```

```python
#univariant boxplot:
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='parental level of education', palette='pastel')
plt.title("parental level of education")
plt.show()
```

```python
#univariant histplot:
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='parental level of education', palette='pastel')
plt.title("parential level of education")
plt.show()
```

```python
#bivariant countplot:
plt.figure(figsize=(12, 6)) 
sns.countplot(data=data, x='gender', hue='math score', palette='pastel')
plt.yticks(rotation=45)
plt.title("score")
plt.show()
```

```python
#bivariant Scatterplot:
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='writing score', y='reading score', hue='gender', palette='pastel')
plt.title("Writing Score vs Reading Score")
plt.show()
```

```python
#bivariant boxplot:
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='gender', y='math score', palette='pastel')
plt.title("Math Score by Gender")
plt.show()
```

```python
#bivariant Barplot
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='gender', y='math score',  palette='pastel')
plt.title("Math Score by Gender")
plt.show()
```

```python
#bivariant line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='gender', y='math score', palette='pastel')
plt.title("Math Score by Gender")
plt.show()
```

```python
#bivariant heatmap:
plt.figure(figsize=(12, 6))
sns.heatmap(pd.crosstab(data['gender'], data['race/ethnicity']),annot=True)
plt.title("race/ethnicity vs gender")
plt.show()
```

```python
#trivariant scatterplot

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='math score', y='reading score', hue='gender', palette='pastel', s=100)
plt.title("Math Score vs Reading Score by Gender", fontsize=14)
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```

```python
#trivariant boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='gender', y='writing score', hue='test preparation course', palette='Set2')

plt.title("Writing Score by Gender & Test Preparation Course", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Writing Score")
plt.legend(title='Test Preparation')
plt.grid(True)
plt.show()
```

```python
#multivariant stacked  bar plot
plt.figure(figsize=(12, 6))
stack_data = data.groupby(['gender', 'test preparation course']).size().unstack()
stack_data.plot(kind='bar', stacked=True)
plt.title("Gender vs Test Prep (Stacked)")
plt.ylabel("Count")
plt.show()
```
